```python
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup
import re
```

### 1. Создадим программу web-scraping с использованием библиотеки BeautifulSoup.

```python
html = requests.get('https://quotes.toscrape.com/')
bs = BeautifulSoup(html.text, 'html.parser')
bs
```

### 2. Реализуем HTML разбор текста.

```python
quotes = bs.find_all('span', {'class' : 'text'})
authors = bs.find_all('small', {'class' : 'author'})
tags = bs.find_all('div', {'class' : 'tags'})
```

```python
table = {'quote' : [], 'author' : [], 'tags' : []}
```

```python
for i in range(0, len(quotes)):
    table['quote'].append(quotes[i].text.strip())
    table['author'].append(authors[i].text.strip())
    table['tags'].append(tags[i].text.strip()[6:].split())
```

### 3. Выведим результат в DataFarme: Цитата | Автор |  Список тегов

```python
df = pd.DataFrame(data = table)
df
```
